```python
#Normally, here I would filter out columns, but there are only 14 columns, and each column matters
#Columns describe height, leafs, photosynthesis, roots, etc.
plants.columns
```

```python
plants_corr = plants[['ACHP', 'PHR', 'AWWGV', 'ALAP', 'ANPL', 'ARD', 'ADWR',
       'PDMVG', 'ARL', 'AWWR', 'ADWV', 'PDMRG']].corr()
sns.heatmap(plants_corr, annot=True)
#I'm seeing which columns have strong coorelations
```

```python
sns.lmplot(x='AWWGV', y='ACHP', data=plants, fit_reg=True, ci=None)
#This shows that the strongest sole factor of photosynthesis activity is wet weight of vegetative growth
```

```python
sns.lmplot(x='ARL', y='ACHP', data=plants, fit_reg=True, ci=None)
#From the surface, this shows that excessive root length lowers the photosynthesis activity
```

```python
sns.lmplot(x='PHR', y='ALAP', data=plants, fit_reg=True, ci=None)
#This shows that the taller the plant, the more leaf area there is
```
